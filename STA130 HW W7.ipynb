1. Simple Linear Regression vs. Multiple Linear Regression

Simple Linear Regression

Definition: Models the relationship between a single predictor (independent variable) and an outcome (dependent variable).
Equation:
Outcome = Œ≤0 + Œ≤1(Predictor) + Œµ
Œ≤0: Intercept
Œ≤1: Slope coefficient for the predictor
Œµ: Error term


Multiple Linear Regression

Definition: Models the relationship between two or more predictors and an outcome.
Equation:
Outcome = Œ≤0 + Œ≤1(Predictor1) + Œ≤2(Predictor2) + ... + Œ≤n(Predictorn) + Œµ
Œ≤0: Intercept
Œ≤1, Œ≤2, ..., Œ≤n: Coefficients for each predictor
Œµ: Error term
Benefit of Multiple Over Simple Linear Regression

Enhanced explanatory power: By incorporating multiple predictors, the model can account for more variability in the outcome.
Control for confounding variables: Helps isolate the effect of each predictor by controlling for others.
2. Continuous Variable vs. Indicator Variable in Simple Linear Regression
Using a Continuous Variable

Definition: Predictor is a quantitative variable that can take an infinite number of values within a range.
Equation:
Outcome = Œ≤0 + Œ≤1(Continuous Predictor) + Œµ
Interpretation: For each unit increase in the continuous predictor, the outcome changes by Œ≤1 units.
Using an Indicator Variable

Definition: Predictor is a binary variable (0 or 1) representing the presence or absence of a categorical effect.
Equation:
Outcome = Œ≤0 + Œ≤1 √ó 1(Indicator Variable) + Œµ
1(Indicator Variable): Indicator function equals 1 if the condition is met and 0 otherwise.
Interpretation: Œ≤0 represents the expected outcome when the indicator is 0. When the indicator is 1, the outcome increases by Œ≤1.
3. Introducing an Indicator Variable into a Multiple Linear Regression
Simple Linear Regression (Only Continuous Predictor)
Outcome = Œ≤0 + Œ≤1(Continuous Predictor) + Œµ

Multiple Linear Regression (Continuous Predictor + Indicator Variable)
Outcome = Œ≤0 + Œ≤1(Continuous Predictor) + Œ≤2 √ó 1(Indicator Variable) + Œµ

Baseline Group: When the indicator is 0, the model is based only on the continuous predictor.
Indicator Group: When the indicator is 1, an additional fixed effect (Œ≤2) is included.


4. Adding an Interaction Between a Continuous and an Indicator Variable in Multiple Linear Regression
Multiple Linear Regression with Interaction Term
Outcome = Œ≤0 + Œ≤1(Continuous Predictor) + Œ≤2 √ó 1(Indicator Variable) + Œ≤3[(Continuous Predictor) √ó 1(Indicator Variable)] + Œµ

Effect of Interaction (Œ≤3):
Definition: Allows the slope of the continuous predictor to differ depending on the category represented by the indicator variable.
Interpretation: Effect of the continuous predictor is Œ≤1 when the indicator is 0 and Œ≤1 + Œ≤3 when the indicator is 1.

1. Identifying Outcome and Predictor Variables

Outcome Variable

Definition: The metric used to measure the effectiveness of the advertising campaigns.
Examples:
Sales Revenue: Total sales generated.
Conversion Rate: Percentage of viewers/customers who make a purchase.
Brand Awareness Score: Survey-based metric indicating brand recognition.
For this explanation, we‚Äôll assume the Outcome Variable is Sales Revenue.

Predictor Variables

Predictor 2: Online Advertising Budget (Continuous Variable)
2. Considering Interaction Effects

What is an Interaction Effect?

An interaction effect occurs when the effect of one predictor variable on the outcome depends on the level of another predictor variable. In this context:

Hypothesis: The effectiveness of TV advertising (TV Budget) may vary depending on how much is spent on online advertising (Online Budget), and vice versa.
Why Consider Interaction?

Synergistic Effects: Combining TV and online advertising might lead to greater effectiveness than the sum of their individual effects.
Diminishing Returns: High expenditure on one medium might reduce the marginal effectiveness of the other.
3. Formulating Regression Models

a. Simple Linear Regression Models (Without Interaction)

TV Advertising Only:
Sales = Œ≤0 + Œ≤1(TV Budget) + Œµ

Œ≤0: Intercept (Base sales when TV Budget is $0)
Œ≤1: Effect of each additional dollar spent on TV advertising
Online Advertising Only:
Sales = Œ≤0 + Œ≤1(Online Budget) + Œµ

Œ≤1: Effect of each additional dollar spent on online advertising
b. Multiple Linear Regression Model (Without Interaction)

Sales = Œ≤0 + Œ≤1(TV Budget) + Œ≤2(Online Budget) + Œµ

Œ≤1: Change in sales for each additional dollar spent on TV, holding Online Budget constant.
Œ≤2: Change in sales for each additional dollar spent on Online, holding TV Budget constant.
c. Multiple Linear Regression Model (With Interaction)

Sales = Œ≤0 + Œ≤1(TV Budget) + Œ≤2(Online Budget) + Œ≤3(TV Budget √ó Online Budget) + Œµ

Œ≤3: Interaction coefficient representing how the effect of TV Budget on Sales changes with Online Budget.
4. Using the Regression Models for Predictions

a. Without Interaction

Equation:
Sales = Œ≤0 + Œ≤1(TV Budget) + Œ≤2(Online Budget)

Prediction Process:

Input: Specify amounts for TV and Online Budgets.
Calculate: Multiply each budget by its respective coefficient and add the intercept.
Output: Predicted Sales Revenue.
Example:

TV Budget: $10,000
Online Budget: $5,000
Coefficients: Œ≤0 = 50,000, Œ≤1 = 3, Œ≤2 = 2
Sales = 50,000 + (3 √ó 10,000) + (2 √ó 5,000) = 50,000 + 30,000 + 10,000 = $90,000

b. With Interaction

Equation:
Sales = Œ≤0 + Œ≤1(TV Budget) + Œ≤2(Online Budget) + Œ≤3(TV Budget √ó Online Budget)

Prediction Process:

Input: Specify amounts for TV and Online Budgets.
Calculate:
Multiply each budget by its respective coefficient.
Compute the interaction term by multiplying TV Budget and Online Budget, then multiply by Œ≤3.
Add all terms with the intercept.
Output: Predicted Sales Revenue.
Example:

TV Budget: $10,000
Online Budget: $5,000
Coefficients: Œ≤0 = 50,000, Œ≤1 = 3, Œ≤2 = 2, Œ≤3 = 0.0001
Sales = 50,000 + (3 √ó 10,000) + (2 √ó 5,000) + (0.0001 √ó 10,000 √ó 5,000) = 50,000 + 30,000 + 10,000 + 5,000 = $95,000

R-squared (R¬≤)
Definition: R¬≤ represents the proportion of variance in the dependent variable explained by the independent variables.
Formula:
R¬≤ = 1 - (Œ£(y·µ¢ - ≈∑·µ¢)¬≤ / Œ£(y·µ¢ - »≥)¬≤)

y·µ¢: Actual outcome.

≈∑·µ¢: Predicted outcome from the model.

»≥: Mean of the actual outcomes.

Interpretation:

R¬≤ = 0: Model explains none of the variability in the outcome.
R¬≤ = 1: Model perfectly explains the variability in the outcome.
Low R¬≤ (e.g., 0.176): Model explains only 17.6% of the variability in the outcome.
P-values and Coefficient Estimates
Coefficient (Œ≤): Represents the change in the dependent variable for a one-unit change in the predictor, holding all other predictors constant.
P-value:
Definition: The probability of observing the estimated coefficient (or something more extreme) if the null hypothesis (Œ≤ = 0) is true.
Interpretation:
Low p-value (< 0.05): Strong evidence against the null hypothesis, indicating the predictor is statistically significant.
High p-value: Weak evidence against the null hypothesis, suggesting the predictor may not be significant.
Explaining the Apparent Contradiction
Scenario:

Model Fit: R¬≤ = 0.176 (17.6% of variability explained).
Coefficient Estimates: Many coefficients > 10 with low p-values.
Apparent Contradiction:
Why is R¬≤ low if predictors are significant?

Resolution:
R¬≤ and p-values assess different aspects of the regression model. Understanding their distinction resolves the contradiction.

R¬≤ vs. P-values: Different Objectives
R¬≤ focuses on the collective explanatory power of all predictors.
P-values focus on the individual significance of each predictor.
Scenarios Where Low R¬≤ and Significant Coefficients Co-occur
Small but Consistent Effects: Predictors have small individual effects that are statistically significant but don‚Äôt explain much overall variability.
High Outcome Variability: The dependent variable has inherent variability not captured by the predictors.
Number of Predictors vs. Observations: Overfitting with many predictors can inflate coefficients‚Äô significance without improving R¬≤.
Multicollinearity: High correlation between predictors can inflate p-values but reduce model efficiency.

1. Understanding the Design Matrix and Predictor Creation
a. What is a Design Matrix?
In regression analysis, the design matrix (often denoted as X) is a structured matrix that contains all the predictor variables (independent variables) used to predict the outcome variable (dependent variable). Each column in the design matrix represents a predictor, and each row corresponds to an observation.

b. How model4_linear_form Creates New Predictors

Centering and Scaling:

center() subtracts the mean from each predictor, centering it around zero.
scale() divides each predictor by its standard deviation, standardizing the scale.
Purpose: Helps in mitigating multicollinearity and ensures that interaction terms are on comparable scales.
Interaction Terms (* Operator):

The * operator in patsy (used by statsmodels) includes both the main effects and all possible interaction terms between the specified predictors.
For example, Attack * Defense expands to:
Attack+Defense+(Attack√óDefense)
When multiple * operators are used, the number of interaction terms increases exponentially.

Categorical Variable (Legendary):

Legendary is treated as a binary indicator variable (e.g., 1 for legendary Pok√©mon, 0 otherwise).
Note: It‚Äôs not centered or scaled because it's already binary.
Implications:
Exponential Growth of Predictors:

Each additional predictor and interaction increases the number of columns in the design matrix.
For six predictors involved in multiple interactions, the number of columns can become extremely large.
Creation of Highly Correlated Predictors:

Interaction terms are products of existing predictors, leading to highly correlated columns in the design matrix.

2. Multicollinearity in the Design Matrix
a. What is Multicollinearity?
Multicollinearity refers to the situation where two or more predictor variables in a regression model are highly correlated. This means that one predictor can be linearly predicted from the others with a substantial degree of accuracy.

b. Detecting Multicollinearity:
Condition Number:
A high condition number (generally above 30) indicates strong multicollinearity.
In your models:
Model 3 (Simple): Condition Number ‚âà 343.0
Model 4 (Complex): Condition Number ‚âà 1.2e+16
Correlation Matrix (np.corrcoef(model4_spec.exog)):
Shows the pairwise correlations between all predictors.
In Model 4, due to numerous interaction terms, many predictors are highly correlated.
c. Consequences of Multicollinearity:
Inflated Variance of Coefficients:
Estimates of regression coefficients become unstable and sensitive to minor changes in the model or data.
Difficulty in Assessing Predictor Importance:
High multicollinearity makes it challenging to determine the individual effect of each predictor.
Overfitting:
The model may fit the training data too closely, capturing noise rather than the underlying relationship, leading to poor generalization on new data.
3. Impact on Out-of-Sample Generalization
a. In-Sample vs. Out-of-Sample Performance:
In-Sample R2 :Measures how well the model explains variability in the training data.
Model 4 likely shows a high in-sample R2 due to the numerous predictors and interactions capturing complex patterns in the training set.

Out-of-Sample R2: Measures how well the model predicts outcomes in new, unseen data.
Model 4 shows a low or even undefined out-of-sample R2 , indicating poor predictive performance.

b. Why Multicollinearity Hurts Generalization:
Overfitting:

The model captures specific patterns (including noise) present only in the training data.
These patterns do not generalize to new data, resulting in poor out-of-sample performance.
Unstable Coefficients:

Highly correlated predictors lead to large and erratic coefficient estimates.
Small changes in data can cause substantial shifts in these estimates, making the model unreliable on new data.
Redundancy:

Multiple predictors conveying similar information make the model unnecessarily complex without adding predictive power.

````python
# Question 8

import pandas as pd
import numpy as np
import statsmodels.formula.api as smf
from sklearn.model_selection import train_test_split
import plotly.express as px
import plotly.graph_objects as go

# Load the Pok√©mon dataset
url = "https://raw.githubusercontent.com/KeithGalli/pandas/master/pokemon_data.csv"
pokeaman = pd.read_csv(url)

# Replace "NaN" in the "Type 2" column with "None"
pokeaman.fillna('None', inplace=True)

# Define the linear model specification for Model 3
model3_linear_form = 'HP ~ Attack + Defense'

# Number of repetitions
reps = 100

# Initialize arrays to store R-squared values
in_sample_Rsquared = np.zeros(reps)
out_of_sample_Rsquared = np.zeros(reps)

for i in range(reps):
    # Split the data into 50% training and 50% testing without a fixed seed
    pokeaman_train, pokeaman_test = train_test_split(pokeaman, train_size=0.5)
    
    # Fit the linear regression model on the training data
    model_spec3 = smf.ols(formula=model3_linear_form, data=pokeaman_train)
    model3_fit = model_spec3.fit()
    
    # Calculate in-sample R-squared
    in_sample_Rsquared[i] = model3_fit.rsquared
    
    # Predict on the test set
    yhat_model3 = model3_fit.predict(pokeaman_test)
    
    # Actual HP values from the test set
    y = pokeaman_test.HP
    
    # Calculate out-of-sample R-squared as the squared correlation between actual and predicted HP
    out_of_sample_Rsquared[i] = np.corrcoef(y, yhat_model3)[0,1]**2

# Create a DataFrame to hold the R-squared values
df_Rsquared = pd.DataFrame({
    "In Sample R-squared": in_sample_Rsquared,
    "Out of Sample R-squared": out_of_sample_Rsquared
})

# Create a scatter plot of In-Sample vs Out-of-Sample R-squared
fig = px.scatter(df_Rsquared, 
                 x="In Sample R-squared", 
                 y="Out of Sample R-squared",
                 title="In-Sample vs Out-of-Sample R-squared Across Multiple Splits",
                 labels={"In Sample R-squared": "In-Sample R¬≤", "Out of Sample R-squared": "Out-of-Sample R¬≤"},
                 opacity=0.6)

# Add a y=x reference line
fig.add_trace(go.Scatter(
    x=[df_Rsquared["In Sample R-squared"].min(), df_Rsquared["In Sample R-squared"].max()],
    y=[df_Rsquared["In Sample R-squared"].min(), df_Rsquared["In Sample R-squared"].max()],
    mode='lines',
    name='y = x',
    line=dict(color='red', dash='dash')
))

# Show the plot
fig.show(renderer="png")
````
