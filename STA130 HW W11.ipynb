### **(a) Understanding the Problem Addressed by Classification Decision Trees and Their Real-World Applications**

A **Classification Decision Tree** is designed to solve **classification problems**, which involve categorizing data into predefined classes or groups. Specifically, it helps in determining the category or class to which a new observation belongs based on various input features.

**Key Characteristics:**
- **Supervised Learning:** It requires labeled data to learn from.
- **Discrete Output:** The output is categorical (e.g., yes/no, spam/ham).
- **Hierarchical Decision-Making:** It makes decisions in a tree-like model of sequential splits based on feature values.


Here are some practical scenarios where Classification Decision Trees excel:

- **Medical Diagnosis:**
  - **Example:** Determining whether a patient has a particular disease (e.g., diabetes) based on symptoms, age, BMI, and other health indicators.
  
- **Customer Segmentation:**
  - **Example:** Classifying customers into groups like "high-value," "medium-value," and "low-value" based on purchasing behavior, demographics, and engagement levels.
  
- **Fraud Detection:**
  - **Example:** Identifying fraudulent credit card transactions by analyzing transaction patterns, location, amount, and frequency.
  
- **Loan Approval:**
  - **Example:** Deciding whether to approve or reject a loan application based on factors like credit score, income, employment history, and existing debts.
  
- **Email Filtering:**
  - **Example:** Classifying emails as "spam" or "not spam" by analyzing content, sender information, and other metadata.
  
- **Image Recognition:**
  - **Example:** Categorizing images into different classes such as "cats," "dogs," "vehicles," etc., based on pixel data and features extracted from images.

These applications benefit from the interpretability and simplicity of decision trees, allowing stakeholders to understand the decision-making process clearly.

### **(b) Difference Between Classification Decision Trees and Multiple Linear Regression in Making Predictions**

While both **Classification Decision Trees** and **Multiple Linear Regression** are supervised learning algorithms, they serve different purposes and operate differently.

1. **_Nature of the Problem They Address_**

- **Classification Decision Trees:**
  - **Purpose:** Used for **classification** tasks where the goal is to assign inputs to discrete categories or classes.
  - **Output:** Categorical labels (e.g., "yes" or "no," "spam" or "not spam").

- **Multiple Linear Regression:**
  - **Purpose:** Used for **regression** tasks where the goal is to predict continuous numerical values.
  - **Output:** Continuous numerical values (e.g., predicting house prices, temperature forecasts).

2. **_How They Make Predictions_**

- **Classification Decision Trees:**
  - **Structure:** Composed of nodes and branches forming a tree-like structure.
  - **Decision Process:**
    - **Sequential Splits:** Starting from the root node, the algorithm makes a series of binary or multi-way splits based on feature values.
    - **Path Traversal:** Each internal node represents a decision based on a specific feature, leading to child nodes.
    - **Leaf Nodes:** The final nodes (leaves) represent the predicted class labels.
  - **Example:**
    - **Root Node:** Is the customer's income > $50K?
      - **Yes:** Go to the next node.
      - **No:** Assign to class "Low Income."

- **Multiple Linear Regression:**
  - **Mathematical Model:** Establishes a linear relationship between input features and the target variable.
  - **Prediction Formula:**
    \[
    \text{Prediction} = \beta_0 + \beta_1X_1 + \beta_2X_2 + \dots + \beta_nX_n
    \]
    where \(\beta\)s are coefficients and \(X\)s are input features.
  - **Interpretation:** Each coefficient represents the weight or influence of a corresponding feature on the prediction.
  - **Example:**
    - **House Price Prediction:** 
      \[
      \text{Price} = 50000 + 30000(\text{Size in Sq Ft}) + 15000(\text{Number of Bedrooms})
      \]

3. **_Key Differences in Prediction Mechanisms_**

- **Output Type:**
  - **Decision Trees:** Categorical (classification).
  - **Linear Regression:** Continuous (regression).

- **Model Interpretability:**
  - **Decision Trees:** Highly interpretable; easy to visualize the decision path.
  - **Linear Regression:** Interpretability is based on understanding the coefficients' impact on the outcome.

- **Handling Feature Interactions:**
  - **Decision Trees:** Naturally capture interactions between features through the tree structure.
  - **Linear Regression:** Requires explicit interaction terms to model feature interactions.

- **Complexity and Flexibility:**
  - **Decision Trees:** Can model non-linear relationships without requiring transformation of features.
  - **Linear Regression:** Assumes a linear relationship between features and the target variable.

4. **_Sequential Decision-Making vs. Linear Combination_**

- **Decision Trees:** Make **sequential decisions** by navigating through the tree's nodes, each representing a condition based on a feature. This hierarchical approach allows the model to handle complex, non-linear relationships by partitioning the feature space into distinct regions.

- **Multiple Linear Regression:** Makes predictions based on a **linear combination** of input features. Each feature contributes additively to the final prediction, assuming a straight-line relationship between each feature and the target variable.

1. Accuracy​
 
Real-World Application: Quality Control in Manufacturing

Scenario: A manufacturing company produces thousands of widgets daily. Ensuring that these widgets meet quality standards is critical to maintaining customer satisfaction and minimizing costs associated with defective products.

Why Accuracy is Appropriate:

Balanced Classes: In this context, the number of good widgets (True Negatives) and defective widgets (True Positives) is relatively balanced.
Overall Performance: Accuracy provides a straightforward measure of how well the quality control system is performing in correctly identifying both defective and non-defective widgets.
Cost Implications: Since both false positives (incorrectly flagging a good widget as defective) and false negatives (failing to identify a defective widget) can have cost implications, a high accuracy ensures that the majority of widgets are correctly classified.
Rationale: Accuracy gives a clear picture of the overall effectiveness of the quality control process when the classes are balanced. It helps the company monitor and maintain the desired level of product quality.

2. Sensitivity
 
Real-World Application: Medical Screening for a Serious Disease

Scenario: A hospital implements a screening test for a serious but treatable disease, such as certain types of cancer. Early detection is crucial for effective treatment and improving patient survival rates.

Why Sensitivity is Appropriate:

Minimizing Missed Cases: High sensitivity ensures that most patients who have the disease are correctly identified (True Positives), reducing the number of cases that go undetected (False Negatives).
Critical for Treatment: Missing a diagnosis can have severe consequences for the patient's health. Therefore, prioritizing sensitivity helps in capturing as many true cases as possible.
Resource Allocation: While high sensitivity may lead to more false positives, the priority is to ensure that those who need treatment are not overlooked.
Rationale: In medical diagnostics, especially for serious conditions, it's paramount to identify as many true cases as possible. Sensitivity focuses on minimizing the chances of missing a diagnosis, which is critical for patient outcomes.

3. Specificity
 
Real-World Application: Spam Email Filtering

Scenario: An email service provider uses a spam filter to classify incoming emails as "spam" or "not spam." Users rely on the filter to keep their inboxes free from unwanted messages.

Why Specificity is Appropriate:

Avoiding False Alarms: High specificity ensures that legitimate emails (True Negatives) are not incorrectly marked as spam (False Positives), which could result in users missing important communications.
User Trust: Maintaining user trust in the email service is essential. If too many legitimate emails are flagged as spam, users may become frustrated and lose confidence in the system.
User Experience: A high specificity minimizes disruptions to the user's normal email flow, ensuring that the primary function of email communication remains uninterrupted.
Rationale: While it's important to filter out spam, it's equally crucial to ensure that legitimate emails are not mistakenly classified as spam. Specificity focuses on correctly identifying non-spam emails, maintaining a balance between filtering unwanted messages and preserving important communications.

4. Precision

Real-World Application: Targeted Advertising in Digital Marketing

Scenario: A digital marketing firm uses an algorithm to identify and target potential customers who are likely to purchase a new product. The algorithm predicts which users are interested based on their online behavior and preferences.

Why Precision is Appropriate:

Maximizing Relevance: High precision ensures that the majority of targeted advertisements reach users who are genuinely interested in the product (True Positives), enhancing the effectiveness of the campaign.
Cost Efficiency: Reducing false positives (users who are not interested but are targeted) minimizes wasted advertising spend on uninterested audiences.
User Engagement: Delivering relevant ads increases user engagement and the likelihood of conversions, improving the overall return on investment for the marketing campaign.
Rationale: In targeted advertising, it's essential to reach the right audience to maximize the impact of the campaign. Precision ensures that the advertisements are shown to users who are more likely to respond positively, making the marketing efforts more efficient and effective.

````python
Questino 5

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score

# Set seaborn style for better visualization
sns.set(style="whitegrid")

# Load the Dataset

url = "https://raw.githubusercontent.com/pointOfive/STA130_F23/main/Data/amazonbooks.csv"

try:
    ab = pd.read_csv(url, encoding="ISO-8859-1")
    print("Dataset loaded successfully.\n")
except Exception as e:
    print(f"Error loading dataset: {e}")

# Pre-process the Dataset

# Remove unnecessary columns
columns_to_remove = ['Weight_oz', 'Width', 'Height']
existing_columns_to_remove = [col for col in columns_to_remove if col in ab.columns]

if existing_columns_to_remove:
    ab_reduced = ab.drop(columns=existing_columns_to_remove)
    print(f"Removed columns: {existing_columns_to_remove}\n")
else:
    ab_reduced = ab.copy()
    print("No specified columns found to remove.\n")

# Drop rows with any NaN values
initial_row_count = ab_reduced.shape[0]
ab_reduced_noNaN = ab_reduced.dropna()
final_row_count = ab_reduced_noNaN.shape[0]
rows_dropped = initial_row_count - final_row_count
print(f"Number of Rows Before Dropping NaNs: {initial_row_count}")
print(f"Number of Rows After Dropping NaNs: {final_row_count}")
print(f"Number of Rows Dropped: {rows_dropped}\n")

# Set Appropriate Data Types using .loc to avoid SettingWithCopyWarning
ab_reduced_noNaN.loc[:, 'Pub year'] = pd.to_numeric(ab_reduced_noNaN['Pub year'], errors='coerce').astype('Int64')
ab_reduced_noNaN.loc[:, 'NumPages'] = pd.to_numeric(ab_reduced_noNaN['NumPages'], errors='coerce').astype('Int64')
ab_reduced_noNaN.loc[:, 'Hard_or_Paper'] = ab_reduced_noNaN['Hard_or_Paper'].astype('category')

# Verify the data types
print("Data Types After Conversion:")
print(ab_reduced_noNaN.dtypes, "\n")

# Display a preview of the preprocessed DataFrame
print("Preprocessed DataFrame Preview:")
print(ab_reduced_noNaN.head(), "\n")

# Define Features and Target Variable for Both Classifiers

# Check if 'List Price' and 'Thick' exist in the DataFrame
required_features_clf2 = ['NumPages', 'Thick', 'List Price']
missing_features_clf2 = [feature for feature in required_features_clf2 if feature not in ab_reduced_noNaN.columns]

if missing_features_clf2:
    raise ValueError(f"The following features are missing from the dataset: {missing_features_clf2}")

# Features for clf: 'List Price'
# Features for clf2: 'NumPages', 'Thick', 'List Price'

X_clf = ab_reduced_noNaN[['List Price']]
X_clf2 = ab_reduced_noNaN[required_features_clf2]

# Target variable: 'Hard_or_Paper' mapped to binary (1 for 'H', 0 for 'P')
y = ab_reduced_noNaN['Hard_or_Paper'].map({'H': 1, 'P': 0})

# Verify mapping
print("Value counts for y (1: H, 0: P):")
print(y.value_counts(), "\n")

# Perform an 80/20 Train-Test Split for Both Classifiers
random_seed = 42

# Split for clf
X_train_clf, X_test_clf, y_train_clf, y_test_clf = train_test_split(
    X_clf, y, test_size=0.2, random_state=random_seed, stratify=y
)

# Split for clf2
X_train_clf2, X_test_clf2, y_train_clf2, y_test_clf2 = train_test_split(
    X_clf2, y, test_size=0.2, random_state=random_seed, stratify=y
)

print(f"Number of observations in the training set for clf: {X_train_clf.shape[0]}")
print(f"Number of observations in the testing set for clf: {X_test_clf.shape[0]}\n")

print(f"Number of observations in the training set for clf2: {X_train_clf2.shape[0]}")
print(f"Number of observations in the testing set for clf2: {X_test_clf2.shape[0]}\n")

# Train Decision Tree Classifier clf (using 'List Price' only)

clf = DecisionTreeClassifier(max_depth=2, random_state=random_seed)
clf.fit(X_train_clf, y_train_clf)

print("Decision Tree Classifier clf has been trained.\n")

# Train Decision Tree Classifier clf2 (using 'NumPages', 'Thick', 'List Price')

clf2 = DecisionTreeClassifier(max_depth=4, random_state=random_seed)
clf2.fit(X_train_clf2, y_train_clf2)

print("Decision Tree Classifier clf2 has been trained.\n")

# Visualize the Decision Tree clf

plt.figure(figsize=(12,8))
plot_tree(clf, feature_names=['List Price'], class_names=['P', 'H'], filled=True, rounded=True)
plt.title('Decision Tree Classifier clf (List Price Only)')
plt.show()

# Visualize the Decision Tree clf2

plt.figure(figsize=(16,12))
plot_tree(clf2, feature_names=required_features_clf2, class_names=['P', 'H'], filled=True, rounded=True)
plt.title('Decision Tree Classifier clf2 (NumPages, Thick, List Price)')
plt.show()

# Function to Plot Confusion Matrix
def plot_confusion_matrix(cm, classes, title):
    plt.figure(figsize=(6,5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=classes, yticklabels=classes)
    plt.ylabel('Actual')
    plt.xlabel('Predicted')
    plt.title(title)
    plt.show()

# Evaluate clf on the Test Set

y_pred_clf = clf.predict(X_test_clf)

accuracy_clf = accuracy_score(y_test_clf, y_pred_clf)
conf_matrix_clf = confusion_matrix(y_test_clf, y_pred_clf)

print(f"Accuracy of clf on Test Set: {accuracy_clf:.3f}")

# Extract TP, TN, FP, FN for clf
if conf_matrix_clf.shape == (2,2):
    TN_clf, FP_clf, FN_clf, TP_clf = conf_matrix_clf.ravel()
    
    sensitivity_clf = TP_clf / (TP_clf + FN_clf)
    specificity_clf = TN_clf / (TN_clf + FP_clf)
    
    print(f"Sensitivity (Recall) of clf: {sensitivity_clf:.3f}")
    print(f"Specificity of clf: {specificity_clf:.3f}\n")
else:
    print("Confusion matrix for clf is not 2x2. Check the data and predictions.\n")

# Plot Confusion Matrix for clf
plot_confusion_matrix(conf_matrix_clf, classes=['P', 'H'], title='Confusion Matrix for clf')

# Evaluate clf2 on the Test Set

y_pred_clf2 = clf2.predict(X_test_clf2)

accuracy_clf2 = accuracy_score(y_test_clf2, y_pred_clf2)
conf_matrix_clf2 = confusion_matrix(y_test_clf2, y_pred_clf2)

print(f"Accuracy of clf2 on Test Set: {accuracy_clf2:.3f}")

# Extract TP, TN, FP, FN for clf2
if conf_matrix_clf2.shape == (2,2):
    TN_clf2, FP_clf2, FN_clf2, TP_clf2 = conf_matrix_clf2.ravel()
    
    sensitivity_clf2 = TP_clf2 / (TP_clf2 + FN_clf2)
    specificity_clf2 = TN_clf2 / (TN_clf2 + FP_clf2)
    
    print(f"Sensitivity (Recall) of clf2: {sensitivity_clf2:.3f}")
    print(f"Specificity of clf2: {specificity_clf2:.3f}\n")
else:
    print("Confusion matrix for clf2 is not 2x2. Check the data and predictions.\n")

# Plot Confusion Matrix for clf2
plot_confusion_matrix(conf_matrix_clf2, classes=['P', 'H'], title='Confusion Matrix for clf2')

# Detailed Classification Reports

print("Classification Report for clf:")
print(classification_report(y_test_clf, y_pred_clf, target_names=['P', 'H']))

print("Classification Report for clf2:")
print(classification_report(y_test_clf2, y_pred_clf2, target_names=['P', 'H']))
````

Question 9

In linear regression, coefficients quantify the exact change in the dependent variable for a one-unit increase in each predictor, providing a clear direction and magnitude of each feature's impact while holding other variables constant. In contrast, feature importances in decision trees measure the relative contribution of each feature to the model’s ability to split the data effectively, without specifying the direction or precise effect size. Thus, while regression coefficients offer direct interpretability of feature effects, decision tree feature importances indicate overall influence on decision-making without detailing specific relationships.
